```python
import pandas as pd
import bokeh
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
output_notebook()
```

```python
df = pd.read_csv(r'data\person type.xls',sep = '\t',header = None)
```

```python
df = df.iloc[3:,:]
df.columns = ['year', 'cars','t1','t2','motor','bus','uk1','total1','p1','p2','uk2','total2','nan','total','nan']
```

```python
years = list(range(1994,2017))
motor = list(map(int,df['motor']))
cars = list(map(int,df['cars']))
t1 = list(df['t1'])
t2 = list(df['t2'])
trucks = [int(t1[i]) + int(t2[i]) for i in range(len(t1))]
total2 = list(map(int,df['total2']))
total = list(map(int,df['total']))
```

```python
p = figure(plot_width=800, plot_height=400, title="person killed, by type (1994~2016)")

p.line(x=years, y=motor, legend = 'Motorcycles', color="orange", line_width=1)
p.circle(x=years, y=motor, line_color="orange",legend = 'Motorcycles', fill_color="orange", size=5)

p.line(x=years, y=cars, legend = 'cars', color="blue", line_width=1)
p.circle(x=years, y=cars, line_color="blue",legend = 'cars', fill_color="blue", size=5)

p.line(x=years, y=trucks, legend = 'trucks', color="green", line_width=1)
p.circle(x=years, y=trucks, line_color="green",legend = 'trucks', fill_color="green", size=5)

p.line(x=years, y=total2, legend = 'Nonmotorists', color="red", line_width=1)
p.circle(x=years, y=total2, line_color="red",legend = 'Nonmotorists', fill_color="red", size=5)

p.y_range.start = 0
p.y_range.end = 25000
p.legend.location = "top_right"
p.legend.orientation = "horizontal"

p.xaxis.axis_label = 'years'
p.yaxis.axis_label = 'number'

show(p)
```
